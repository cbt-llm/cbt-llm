```python
# Clean all GPT-patient conversations
# Input:  /data/raw/Conv-GPT-patients (eng)-1/
# Output: /data/processed/cleaned-gpt-conv/

import os
import json
from pathlib import Path

BASE_DIR = Path(os.getcwd()).resolve().parent
RAW_DIR = BASE_DIR / "data" / "raw" / "Conv-GPT-patients (eng)-1"
OUT_DIR = BASE_DIR / "data" / "processed" / "cleaned-gpt-conv"

OUT_DIR.mkdir(parents=True, exist_ok=True)

print("BASE_DIR:", BASE_DIR)
print("RAW_DIR:", RAW_DIR)
print("OUT_DIR:", OUT_DIR)


def clean_conversation_file(path: Path):
    with open(path, "r") as f:
        raw = json.load(f)

    cleaned = []

    for entry in raw:
        if "psychotherapist pre-prompt" in entry:
            continue
        if "patient pre-prompt" in entry:
            continue
        if "reminder psychotherapist" in entry:
            continue
        if "reminder patient" in entry:
            continue

        if entry.get("role") in ["user", "assistant"]:
            text = entry.get("content", "").strip()

            if len(text) < 3:
                continue
            if text.startswith("Please play the role"):
                continue
            if "act as a" in text.lower():
                continue

            cleaned.append({
                "role": entry["role"],
                "content": text
            })

    return cleaned

all_cleaned = []

json_files = sorted(RAW_DIR.glob("Conversazione*_GPT-patient.json"))

print(f"Found {len(json_files)} files.")

for path in json_files:
    cleaned = clean_conversation_file(path)

    out_path = OUT_DIR / f"{path.stem.replace('_GPT-patient', '')}_cleaned.json"
    with open(out_path, "w") as f:
        json.dump(cleaned, f, indent=2)

    all_cleaned.extend(cleaned)

    print(f"Cleaned {path.name} → {out_path.name} ({len(cleaned)} messages)")


combined_out_path = OUT_DIR / "cleaned_all_conversations.json"
with open(combined_out_path, "w") as f:
    json.dump(all_cleaned, f, indent=2)


print(f"Total cleaned messages: {len(all_cleaned)}")
print(f"Combined dataset saved to: {combined_out_path}")
```
